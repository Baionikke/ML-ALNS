## CREATION VECTOR WITH IMPROVEMENT FOR EACH ITERATIONS (MEAN)

##### ORIGINAL

```python
from ast import literal_eval
interations_diff_tot_original = zeros(10)
line=0
with open('./EVRPTW-main/original_results.csv', 'r') as originalresultsR:
    readerMR = csv.reader(originalresultsR,delimiter=",")
    for row in readerMR:
        if (line>0):
            array_5 = (literal_eval(row[6]))
            for i in range(10):
                interations_diff_tot_original[i]+=array_5[i]
        line+=1

print("Iteration1,Iteration2,Iteration3...")
print("SUM")
print(interations_diff_tot_original)

interations_diff_mean_original=interations_diff_tot_original
for i in range(10):
    interations_diff_mean_original[i]=interations_diff_tot_original[i]/line
print("AVG")
print(interations_diff_mean_original)
```

##### BINARY MODEL

```python
from ast import literal_eval
interations_diff_tot_binarymodel = zeros(10)
line=0
with open('./EVRPTW-main-modelled-binary/model_results_binary.csv', 'r') as binarymodelresultsR:
    readerMR = csv.reader(binarymodelresultsR,delimiter=",")
    for row in readerMR:
        if (line>0):
            array_5 = (literal_eval(row[6]))
            for i in range(10):
                interations_diff_tot_binarymodel[i]+=array_5[i]
        line+=1

print("Iteration1,Iteration2,Iteration3...")
print("SUM")
print(interations_diff_tot_binarymodel)

interations_diff_mean_binarymodel=interations_diff_tot_binarymodel
for i in range(10):
    interations_diff_mean_binarymodel[i]=interations_diff_tot_binarymodel[i]/line
print("AVG")
print(interations_diff_mean_binarymodel)
```
